```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('D:/ML_Code/realestate.csv')
```

```python
data.head()
```

```python
print(data.shape )
data = data.dropna()
print(data.shape )
```

```python
X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split( X, y, test_size = 1/4, random_state = 0) 
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor = regressor.fit(X_train, Y_train)
```

```python
Y_pred = regressor.predict(X_test)
```

# Training Results

```python
plt.scatter(list(data['No'])[:310] , Y_train, color = 'red')
plt.plot(list(data['No'])[:310] , regressor.predict(X_train), color ='blue')
```

# Testing Results

```python
plt.scatter(list(data['No'])[310:] , Y_test, color = 'red')
plt.plot(list(data['No'])[310:] , regressor.predict(X_test), color ='blue')
```
